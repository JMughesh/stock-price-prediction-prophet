```python
import yfinance as yf

ticker_symbol = "ITC.NS"
stock_data = yf.download(ticker_symbol)
```

```python
display(stock_data.head())
```

```python
display(stock_data.head())
```

```python
close_prices = stock_data['Close']
close_prices = close_prices.reset_index()
close_prices.columns = ['ds', 'y']
print("Missing values in close_prices:\n", close_prices.isnull().sum())
```

```python
from prophet import Prophet

model = Prophet()
```

```python
model.fit(close_prices)
```

```python
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)
display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

merged_df = pd.merge(close_prices, forecast[['ds', 'yhat']], on='ds')

mae = mean_absolute_error(merged_df['y'], merged_df['yhat'])
rmse = np.sqrt(mean_squared_error(merged_df['y'], merged_df['yhat']))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

merged_df = pd.merge(close_prices, forecast[['ds', 'yhat']], on='ds')

mae = mean_absolute_error(merged_df['y'], merged_df['yhat'])
rmse = np.sqrt(mean_squared_error(merged_df['y'], merged_df['yhat']))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(merged_df['ds'], merged_df['y'], label='Actual Close Price')
plt.plot(merged_df['ds'], merged_df['yhat'], label='Predicted Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Stock Price Prediction vs Actual')
plt.legend()
plt.show()
```
